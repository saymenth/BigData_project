```python
# get mean total traffic dari jam 1-12 dan 13-24
pagi_12 = df.iloc[:, 3:15].mean(axis=1)
malam_12 = df.iloc[:, 15:27].mean(axis=1)
```

```python
# drop columns data total traffic dari jam 1-24
df.drop(df.columns[3:33], axis=1, inplace=True)
```

```python
# assign data mean traffic ke dataframe
df['pagi_12'] = pagi_12
df['malam_12'] = malam_12
```
