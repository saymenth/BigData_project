```python
# get mean total traffic dari jam 1-12 dan 13-24
pagi_12 = df.iloc[:, 3:15].mean(axis=1)
malam_12 = df.iloc[:, 15:27].mean(axis=1)
```

```python
# drop columns data total traffic dari jam 1-24
df.drop(df.columns[3:27], axis=1, inplace=True)
```

```python
# assign data mean traffic ke dataframe
df['morning-noon'] = pagi_12
df['noon-night'] = malam_12
```

```python
df.head(3)
```

```python
df.info()
```

```python
col_round = df.columns[3:5]
print(col_round)
```

```python
col_round = df.columns[3:5]
for i in col_round:
    df[i] = np.ceil(df[i].dropna().astype(float))
```

```python
# ubah tipe data date dari string ke datetime -> find yang year 2012 -> tipe data akan datetime jika mau diubah string lagi bolehhh
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
df = df[df['Date'].dt.year == 2012]
```

```python
# Resetting the index and dropping the old index column
df.reset_index(drop=True, inplace=True)
```

```python
df['Roadway Name'] = df['Roadway Name'].astype('category')
```

## Feature Engineering

```python
df_weather['Precip Type'].unique()
```

```python
matching_rows = df_weather.loc[df_weather.isna().any(axis=1)]

# Print the resulting DataFrame
matching_rows.Summary.unique()
```

```python
def fill_precip_type(row):
    if pd.isna(row['Precip Type']):
        if row['Summary'] in ['Mostly Cloudy', 'Partly Cloudy', 'Overcast']:
            return 'Cloudy'
        elif row['Summary'] == 'Clear':
            return 'Clear'
        elif row['Summary'] == 'Foggy':
            return 'Fog'
    return row['Precip Type']

# Apply the function to each row
df_weather['Precip Type'] = df_weather.apply(fill_precip_type, axis=1)

# Optionally, print to verify
print(df_weather['Precip Type'].unique())
```
